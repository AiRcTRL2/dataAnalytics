#### The minimum value column is still skewed since eliminating the duplicate values (since that would leave one column with the -9 value all the way across. We should drop this row in order to see what the minimum values are.

#### After updating to remove the -9 rows below, we notice that more "special meaning values" are visible in the data. This includes -8 but not -7. It's important to note that these features are not intended to represent numerical values, but are coded with meanings. Therefore, if we are looking to analyse the numerical representation of these data, the data cannot be combined catergorical with continuous. 

#### Here, I have made the decision create a second statistical table. One statistical table will be used to analyse the statistics solely of continuous features (i.e. discreet excluded) and the other will contain all information. This is because some of the -8 values could, in theory, be accurate to describe the feature. For example, MSinceMostRecentDelq has minus values. These minus values do not represent what the feature is assessing (which is a count of how long it has been since the entry has experienced a delinquency, i.e. the scale should start from 0. Therefore, we don't want these features in the table to analyse statistics, however they should also be considered useful information since those rows carry special meaning and may provide alternative insights. So for the purposes of statistical analysis, the table used should be the one that does not account for special meaning figures.

```python
# Remove special meaning figures for the stats table (but keep them for alternative areas of research)
# That's why we are creating a new variable instead of updating the old one.
ctDataCleanedfromSpecialMeaningFigs = ctDataCleaned[ctDataCleaned.ExternalRiskEstimate != -9]
ctDataCleanedfromSpecialMeaningFigs = ctDataCleanedfromSpecialMeaningFigs[ctDataCleanedfromSpecialMeaningFigs.MSinceMostRecentDelq != -8]

# Test again to see if it worked
ctDataCleanedfromSpecialMeaningFigs.select_dtypes(['float64', 'int64']).describe().T
```

# 2. Data Quality Plan -- Actions

### Actions to be carried out per Quality Plan document:
-  Remove NaN instances for our target feature RiskPerformance
-  Replace -8 special values with mean where they occur in each column. Each of these instances has information we would not like to lose, and -8 doesn't occur enough to warrant dropping any of the features.
-  Drop the remaining two rows that have -9 special meaning values, since they don't appear to have any useful information in the instance.
-  Drop feature MSinceMostRecentDelq since over ~46% of values are essentially missing or unusable (-7).
-  Impute the mean for entries who have never been delinquent but are also recorded as being delinquent at some stage.

#### We are going to drop the instances where RiskPerformance is NaN, as this is a target feature.

```python
ctDataCleaned.dropna(subset=['RiskPerformance'], inplace=True)
```

```python
# Test if dropping the NaN columns worked.
null_data = ctDataCleaned[ctDataCleaned.isnull().any(axis=1)]
null_data
```

#### Drop rows with all -9 values

```python
# Single line not solid enough, as a single -9 might be present an useful.
#ctDataCleaned.drop(ctDataCleaned[ctDataCleaned.MSinceMostRecentDelq == -9].index, inplace=True)

# Specify the drop on multiple conditions to ensure only rows that are -9 across will be dropped.
ctDataCleaned.drop(ctDataCleaned[(ctDataCleaned['MSinceMostRecentDelq'] == -9) &\
                                 (ctDataCleaned['ExternalRiskEstimate'] == -9) &\
                                 (ctDataCleaned['MSinceOldestTradeOpen'] == -9) &\
                                 (ctDataCleaned['MSinceMostRecentTradeOpen'] == -9) &\
                                 (ctDataCleaned['AverageMInFile'] == -9) &\
                                 (ctDataCleaned['NumTotalTrades'] == -9) &\
                                 (ctDataCleaned['NumInqLast6M'] == -9) &\
                                 (ctDataCleaned['NetFractionRevolvingBurden'] == -9)].index, inplace=True)
```

```python
# test to make sure they're dropped
ctDataCleaned[ctDataCleaned.MSinceMostRecentDelq == -9]
```

#### Turn -8 values into mean in respective columns

```python
# First set -8 to NaN
for i in cleanedContinuousCols:
    ctDataCleaned[i].replace(-8, np.nan,inplace=True)

#Next fill NaN values with the mean of the respective column
for i in cleanedContinuousCols:
    ctDataCleaned[i].fillna((ctDataCleaned[i].mean()), inplace=True)
```

```python
# test to make sure they've been updated
ctDataCleaned[ctDataCleaned.NetFractionInstallBurden == -8]
```
