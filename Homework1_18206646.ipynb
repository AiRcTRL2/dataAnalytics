# 2. Data Quality Plan -- Actions

### Actions to be carried out per Quality Plan document:
-  Remove NaN instances for our target feature RiskPerformance
-  Make -7 = NaN in MSinceMostRecentDelq and MSinceMostRecenetInqexcl7days so it doesn't skew our statistical analysis. Additionally, it would allow us to correlate features between those who do have a value and those who don't at a later stage.
-  Drop the remaining two rows that have -9 special meaning values, since they don't appear to have any useful information in the instance.
-  Set -8 = NaN in NetFractionInstallBurden for the same reason as MSinceMostRecentDelq and MSinceMostRecenetInqexcl7days.
-  Replace -8 special values with mean where they occur in each column with a low percentage of incidence for this value. Each of these instances has information we would not like to lose, and -8 doesn't occur enough to warrant dropping any of the features.
-  Replace PercentTradesNeverDelq == 100 with NaN where MaxDelqEver != 8 (special meaning - current and never delinquent) since this is obviously erroneous.

#### We are going to drop the instances where RiskPerformance is NaN, as this is a target feature.

```python
ctDataCleaned.dropna(subset=['RiskPerformance'], inplace=True)
```

```python
# Test if dropping the NaN columns worked.
null_data = ctDataCleaned[ctDataCleaned.isnull().any(axis=1)]
null_data
```

#### Make -7 in MSinceMostRecentDelq and MSinceMOstRecentInqExcl7days == NaN so later we can correlate features with a value or without a value (effectively acting as a boolean).

```python
ctDataCleaned["MSinceMostRecentInqexcl7days"].replace(-7, np.nan,inplace=True)
ctDataCleaned["MSinceMostRecentDelq"].replace(-7, np.nan,inplace=True)
```

#### Drop rows with all -9 values

```python
# Single line not solid enough, as a single -9 might be present an useful.
#ctDataCleaned.drop(ctDataCleaned[ctDataCleaned.MSinceMostRecentDelq == -9].index, inplace=True)

# Specify the drop on multiple conditions to ensure only rows that are -9 across will be dropped.
ctDataCleaned.drop(ctDataCleaned[(ctDataCleaned['MSinceMostRecentDelq'] == -9) &\
                                 (ctDataCleaned['ExternalRiskEstimate'] == -9) &\
                                 (ctDataCleaned['MSinceOldestTradeOpen'] == -9) &\
                                 (ctDataCleaned['MSinceMostRecentTradeOpen'] == -9) &\
                                 (ctDataCleaned['AverageMInFile'] == -9) &\
                                 (ctDataCleaned['NumTotalTrades'] == -9) &\
                                 (ctDataCleaned['NumInqLast6M'] == -9) &\
                                 (ctDataCleaned['NetFractionRevolvingBurden'] == -9)].index, inplace=True)
```

```python
# test to make sure they're dropped
ctDataCleaned[ctDataCleaned.MSinceMostRecentDelq == -9]
```

#### Set -8 in NetFractionInstallBurden and MSinceMostRecentInqexcl7days == NaN as a flag indicator. Will allow us to continue using remaining data with values in further statistical analysis and feature correlation.

```python
ctDataCleaned["NetFractionInstallBurden"].replace(-8, np.nan,inplace=True)
ctDataCleaned["MSinceMostRecentInqexcl7days"].replace(-8, np.nan,inplace=True)
```

```python
# test for null values
print(ctDataCleaned["NetFractionInstallBurden"].isnull().sum())
print(ctDataCleaned["MSinceMostRecentInqexcl7days"].isnull().sum())
```

#### Turn -8 values into mean in respective columns

```python
# Exclude MSinceMostRecentDelq, MSinceMostRecentInqexcl7days and NetFractionInstallBurden 
# since we want to deal with those separately and -8 doesn't appear in a high percentage
# in any of these other columns, so it's safe to impute the mean in each.

excluded = ["ExternalRiskEstimate",\
              "PercentTradesNeverDelq",\
              "PercentInstallTrades",\
              "NetFractionRevolvingBurden",\
              "NumRevolvingTradesWBalance",\
              "NumInstallTradesWBalance",\
              "NumBank2NatlTradesWHighUtilization",\
              "PercentTradesWBalance",\
              "MSinceOldestTradeOpen",\
              "MSinceMostRecentTradeOpen",\
              "AverageMInFile",\
              "NumSatisfactoryTrades",\
              "NumTrades60Ever2DerogPubRec",\
              "NumTrades90Ever2DerogPubRec",\
              "NumTotalTrades",\
              "NumTradesOpeninLast12M",\
              "NumInqLast6Mexcl7days"]

# First set -8 to NaN
for i in excluded:
    ctDataCleaned[i].replace(-8, np.nan,inplace=True)

#Next fill NaN values with the mean of the respective column
for i in excluded:
    ctDataCleaned[i].fillna((ctDataCleaned[i].mean()), inplace=True)

# Manually replace -8s in MSinceMostRecentDelq since we have many NaN values that we want to keep as NaN as flags 
# and not confuse them with missing data.
ctDataCleaned['MSinceMostRecentDelq'].replace(-8,ctDataCleaned['MSinceMostRecentDelq'].mean(),inplace=True)
```

```python
# test to make sure they've been updated
ctDataCleaned[ctDataCleaned.MSinceMostRecentDelq == -8]
```

#### Replace PercentTradesNeverDelq == 100 with NaN where MaxDelqEver != 8 (special meaning - current and never delinquent) since this is obviously erroneous.

```python
# First check if we have missing values
ctDataCleaned['PercentTradesNeverDelq'].isnull().sum()
```

```python
# check for trades that have been delinquent versus the percentage of accounts listed as having 100% no
# delinquency. Since this it is impossible to be listed as having "PercentTradesNeverDelq=100" and simultaneously
# not having a value of 8 for "MaxDelqEver", this feature needs to be rectified.
ctDataCleaned.loc[(ctDataCleaned["PercentTradesNeverDelq"] == 100) & (ctDataCleaned["MaxDelqEver"] != 8),"PercentTradesNeverDelq"].count()
```

```python
# Convert all values where 100 appears but MaxDelqEver is not 8 or 1 (special meaning values)
# and set it to NaN to signify missing values. These data were considered erroneous, as you cannot have
# 100% never had delinquency while simultaneously having been recorded as having a delinquency.

ctDataCleaned.loc[(ctDataCleaned["PercentTradesNeverDelq"] == 100) & (ctDataCleaned["MaxDelqEver"] != 8), "PercentTradesNeverDelq"]\
                    = np.nan

ctDataCleaned.loc[(ctDataCleaned["PercentTradesNeverDelq"] == 100) & (ctDataCleaned["MaxDelqEver"] != 8), "PercentTradesNeverDelq"]
```

```python
# Check to see if it worked
ctDataCleaned['PercentTradesNeverDelq'].isnull().sum()
```
