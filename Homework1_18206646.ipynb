#### Identifying duplicate rows

```python
pd.set_option('display.max_rows', 25)
# Let's use the df.duplicated() function to identify if we have duplicates. ctData.duplicated() would return both 
# True and False rows. So, we define only to return those rows where the data is in fact duplicated.
ctData[ctData.duplicated() == True]
```

#### Now that we have identified that we have 51 duplicated rows, let's drop them.

## Now we are going to reimport the cleaned CSV and perform some statistical analysis for the data report

```python
ctDataCleaned = pd.read_csv('cleanedCreditRisk-18206645.csv')
ctCategorical = ['RiskPerformance','MaxDelq2PublicRecLast12M','MaxDelqEver']
# Category data does not save to CSV file correctly, so we must convert it again for our new file.
ctDataCleaned['RiskPerformance'] = ctData['RiskPerformance'].astype("category",copy=False)
ctDataCleaned['MaxDelq2PublicRecLast12M'] = ctData['MaxDelq2PublicRecLast12M'].astype("category",copy=False)
ctDataCleaned['MaxDelqEver'] = ctData['MaxDelqEver'].astype("category",copy=False)

ctDataCleaned.dtypes

## Note: Pandas added an unnamed column as a unique ID column, since it was missing from our original data set.
## It automatically assigned this as an integer, which causes us problems when we calculate statistics.
```

```python
# Print table with descriptive statistics for all the continuous features
ctDataCleaned.select_dtypes(['float64', 'int64']).describe().T
```

#### The minimum value column is still skewed since eliminating the duplicate values (since that would leave one column with the -9 value all the way across. We should drop this row in order to see what the minimum values are.

#### After updating to remove the -9 rows below, we notice that more "special meaning values" are visible in the data. This includes -8 but not -7. It's important to note that these features are not intended to represent numerical values, but are coded with meanings. Therefore, if we are looking to analyse the numerical representation of these data, the data cannot be combined catergorical with continuous. 

#### Here, I have made the decision create a second statistical table. One statistical table will be used to analyse the statistics solely of continuous features (i.e. discreet excluded) and the other will contain all information. This is because some of the -8 values could, in theory, be accurate to describe the feature. For example, MSinceMostRecentDelq has minus values. These minus values do not represent what the feature is assessing (which is a count of how long it has been since the entry has experienced a delinquency, i.e. the scale should start from 0. Therefore, we don't want these features in the table to analyse statistics, however they should also be considered useful information since those rows carry special meaning and may provide alternative insights. So for the purposes of statistical analysis, the table used should be the one that does not account for special meaning figures.

```python
# Remove special meaning figures for the stats table (but keep them for alternative areas of research)
# That's why we are creating a new variable instead of updating the old one.
ctDataCleanedfromSpecialMeaningFigs = ctDataCleaned[ctDataCleaned.ExternalRiskEstimate != -9]
ctDataCleanedfromSpecialMeaningFigs = ctDataCleanedfromSpecialMeaningFigs[ctDataCleanedfromSpecialMeaningFigs.MSinceMostRecentDelq != -8]
#ctDataCleanedfromSpecialMeaningFigs = ctDataCleanedfromSpecialMeaningFigs[ctDataCleanedfromSpecialMeaningFigs.MSinceMostRecentDelq != -7]

#ctDataCleanedfromSpecialMeaningFigs = ctDataCleaned.

# Test again to see if it worked
ctDataCleanedfromSpecialMeaningFigs.select_dtypes(['float64', 'int64']).describe().T
```

## Plotting the data

```python
# Plot a histogram for all continuous features
cleanedContinuousCols = ["ExternalRiskEstimate",\
                  "PercentTradesNeverDelq",\
                  "PercentInstallTrades",\
                  "NetFractionRevolvingBurden",\
                  "NetFractionInstallBurden",\
                  "NumRevolvingTradesWBalance",\
                  "NumInstallTradesWBalance",\
                  "NumBank2NatlTradesWHighUtilization",\
                  "PercentTradesWBalance",\
                  "MSinceOldestTradeOpen",\
                  "MSinceMostRecentTradeOpen",\
                  "AverageMInFile",\
                  "NumSatisfactoryTrades",\
                  "NumTrades60Ever2DerogPubRec",\
                  "NumTrades90Ever2DerogPubRec",\
                  "MSinceMostRecentDelq",\
                  "NumTotalTrades",\
                  "NumTradesOpeninLast12M",\
                  "NumInqLast6Mexcl7days"]

ctDataCleaned[cleanedContinuousCols].hist(figsize=(30,30))
plt.savefig('continuous_histograms_1-1.pdf')
```

```python
# Plot box plots for all the continuous features and save in a PDF file
pp = PdfPages('continuous_boxplots_1-1.pdf')

for col in cleanedContinuousCols:
    f = ctDataCleaned[col].plot(kind='box', figsize=(10,5))
    pp.savefig(f.get_figure())
    plt.show()

pp.close()
    
```

```python
# Plot bar charts for all categorical features and save them in a single PDF file
pp = PdfPages('categorical_barcharts_1-1.pdf')

for column in ctCategorical:
    f = ctDataCleaned[column].value_counts().plot(kind='bar', title=column, figsize=(12,10))
    pp.savefig(f.get_figure())
    plt.show()

pp.close()
```
