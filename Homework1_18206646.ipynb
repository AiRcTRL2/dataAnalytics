# Data Quality Report

```python
#Import package pandas for data analysis
import pandas as pd

# Import package numpy for numeric computing
import numpy as np

# Import package matplotlib for visualisation/plotting
import matplotlib.pyplot as plt
```

```python
# Reading credit data from a csv file, into a data frame
ctData = pd.read_csv('CreditRisk-18206646.csv')
# Show data frame for first few rows to ensure everything is working correctly. 
ctData.head()
```

```python
# Use .shape to find how many (rows,columns) exist in the data
ctData.shape
```

```python
# Show the first five rows
ctData[:5] 
```

```python
# Show the last 5 rows
ctData[-5:]
```

```python
# Convert columns we want as continuous values into 
# The values I've chosen here are converted to float as they represent a requirement to have a continuous scale
# among the real number line as they cannot be counted.
# See https://www.dummies.com/education/math/statistics/types-of-statistical-data-numerical-categorical-and-ordinal/
# for more a clarification on this.
continuousCols = ["ExternalRiskEstimate",\
                  "PercentTradesNeverDelq",\
                  "AverageMInFile",\
                  "PercentInstallTrades",\
                  "NetFractionRevolvingBurden",\
                  "NetFractionInstallBurden",\
                  "NumRevolvingTradesWBalance",\
                  "NumInstallTradesWBalance",\
                  "Num2Bank2NatlTradesWHighUtilization",\
                  "PercentTradesWBalance"]
for i in ctData:
    if i in continuousCols:
        ctData[i] = ctData[i].astype("float64",copy=False)

# Categorical data in Pandas requires a limited number of possible values. Thus, RiskPerformance unordered 
# categorical data, while MaxDelq2PublicRecLast12M and MaxDelqEver are the only ordered categorical data.
ctData['RiskPerformance'] = ctData['RiskPerformance'].astype("category",copy=False)
ctData['MaxDelq2PublicRecLast12M'] = ctData['MaxDelq2PublicRecLast12M'].astype("category",copy=False)
ctData['MaxDelqEver'] = ctData['MaxDelqEver'].astype("category",copy=False)

# Show that columns have been converted
ctData.dtypes 


## Note for later: Data doesn't have unique IDs?
```

#### Identifying duplicate rows

```python
pd.set_option('display.max_rows', 25)
# Let's use the df.duplicated() function to identify if we have duplicates. ctData.duplicated() would return both 
# True and False rows. So, we define only to return those rows where the data is in fact duplicated.
ctData[ctData.duplicated() == True]
```

#### Now that we have identified that we have 51 duplicated rows, let's drop them.

```python
# Use the pandas function to drop duplicate rows
ctData.drop_duplicates(inplace=True)
# check again if we have duplicates, and if 0 is returned, we know it worked!
ctData[ctData.duplicated() == True]
```

#### We can see that the duplicate removal worked, since our DataFrame now has 949 rows left (1000-51 duplicates). The next step is to make sure we have no columns with constant values.

```python
# Identify the columns along with their count of unique values
print("Column \t|| UniqueValues") 
for column in ctData:
    print(column + "\t|| " + str(len(ctData[column].unique())))
```

#### Now that we can see that there are no columns in the dataset that has just 1 unique value, we can continue. Let's save our cleaned CSV.

```python
# Save updated DataFrame to CSV
ctData.to_csv('cleanedCreditRisk-18206645.csv')
```
